```python
text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
```

```python
text = "1,23,456,7890,77a8888"
print(re.findall(r"\d{3}", text)) # 숫자문자가 3번연속 나오는 구간만 반환 리스트에 추가
# ['456', '789', '888']
```

```python
#수랑자{2,4}    #앞문자가 최소 n번, 최대m번 반복된다
text = "1 12 123 1234 12345"

print(re.findall(r"\d{2,4}", text)) # ['12', '123', '1234', '1234']
```

```python
# or | 연산 #여러 패턴 중 하나와 일치하면 된다
text = "cat, dog, bird, cats, dog"

print(re.findall(r"cat|dog", text)) # ['cat', 'dog', 'cat', 'dog']
```

```python
phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```

```python
products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]
prices=[]
fruits=[]

for product in products:
    price = re.search(r"\d+", product)
    #print(price.group())
    prices.append(price.group())
for i in products:
    name= re.search(r"\w+",i)
    #print(name.group())
    fruits.append({name.group():int(price.group())})
    
print(fruits)
```

```python
emails = [
    "python@test.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
# + = 숫자 무한 / \. = 마침표(그냥.은 한글자 대신) / {2,} = 두자리수 이상 
for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
#^https? = https 이나 http 로 시작
#:// = 문자열
#([\w-]+\.) = "\w" 이거나 "-" 무제한 , \. 마침표
#(/[^\s]*)?$ = \는 문자 / [^\s]* 공백이 아닌 모든것이 올 수 있다. 0개 이상 / ? 있어도 없어도 된다

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
