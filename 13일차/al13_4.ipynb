```python
# 누적합

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]
```

```python
step_list=[]

for i in steps:
    for value in i.values():
        step_list.append(value)
print(step_list)
```

```python
date_list=[]

for d in steps:
    for key in d.keys():
        date_list.append(key)

print(date_list)
```

```python
#7일간의 걸음수 총합

total=0

for t in range(7):
    total=total + step_list[t]

print(total)
```

```python
#일간 누적 걸음수

daily=[0]

for j in step_list:
    daily.append(daily[-1]+j)
    
print(daily)
```

```python
for i in range(7):
        print(f"{date_list[i]} 누적: {daily[i+1]}")
```

```python
step_list=[]
date_list=[]

for i in steps:
    for value in i.values():
        step_list.append(value)     #print(step_list)   [5231, 6210, 4890, 7005, 6500, 7200, 6800]
    for key in i.keys():
        date_list.append(key)       #print(date_list)   ['2026-01-01', '2026-01-02', '2026-01-03', '2026-01-04', '2026-01-05', '2026-01-06', '2026-01-07']

daily=[0]

for j in step_list:
    daily.append(daily[-1]+j)       #print(daily)       [0, 5231, 11441, 16331, 23336, 29836, 37036, 43836]
    
for i in range(7):
        print(f"{date_list[i]} 누적: {daily[i+1]}")
```
