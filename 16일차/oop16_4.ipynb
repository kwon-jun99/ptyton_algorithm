```python
'''
객체지향의 장점을 이용한 상품 관리 예제
- 속성: 상품명(str), 가격(int), 재고(int)
- 메서드: 주문, 입고, 출고, 할인, 품절 확인
- 객체 상태 변화: 주문과 입고에 따라 재고가 계속 변경됨
- 리팩터링: 딕셔너리 방식에서는 기능이 흩어지지만, 클래스에서는 Product 하나에 기능을 모아 유지보수가 쉬워진다.
'''
```

```python
class Product:
    def __init__(self,name,price,stock): 
        self.name=name
        self.price=price
        self.stock=stock
        
    def total_price(self,count):
        return self.price*count

    def print_info(self):
        print(self.name)
        print(self.price)
        print(self.stock)
    
    def stock_in(self,num1):        #수량이 늘어날때
        return self.stock + num1
    
    def stock_out(self,num2):       #수량이 줄어들때
        return self.stock - num2
    
    def change_price(self,change):  #가격이 바뀔 때
        self.price = change
        return self.price
    
p=Product("키보드",50000,10)
#print(p.name) #print(p.price) #print(p.stock)
p.print_info()
```

```python
#수량이 10개 이상인 물품 찾기

products=[
    Product("키보드",50000,10),
    Product("마우스",30000,20),
    Product("노트북",1200000,3),
    Product("모니터",250000,5)    
]
for p in products:
    if p.stock>=10:
        #p.print_info()
        print(p.name)
```

```python
#세일즈 목록
sales=[
    Product("키보드",50000,10),
    Product("키보드",50000,10),
    Product("마우스",30000,20),
    Product("노트북",1200000,3),
    Product("모니터",250000,5),
    Product("모니터",250000,5),
    Product("모니터",250000,5)    
]

counts={}
for i in sales:
    if i.name not in counts:
        counts[i.name] =1
    else:
        counts[i.name] +=1  #{"키보드":2,"마우스":1,"노트북":1,"모니터":3}

print(counts)

check_list=[]   # 같은 상품이름 중복없이 상품별 판매액 찾기
for s in sales:
    if s.name not in check_list:
        #counts= counts[s.name]
        print(f"{s.name}:{s.total_price(counts[s.name])}"+"원")
        check_list.append(s.name)
```

```python
#5. 객체의 상태변경

print(products[0].stock_in(5))              #키보드 입고 5개
print(products[1].stock_out(3))             #마우스 출고 3개
print(products[3].change_price(270000))     #모니터 가격 270000원으로 변경

```
