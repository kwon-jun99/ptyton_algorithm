```python
n = int(input("크기 : "))

matrix = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

print(matrix)
```

```python
queue = ["A", "B", "C"]

pop_index=0

while pop_index < len(queue):
    print(queue[pop_index])
    print(queue[pop_index+1:])
    pop_index+=1

print(queue)
```

```python
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft())
print(queue)
print(queue.popleft())
print(queue)
```
