```python
nums=[1,3,5,7,9,10,20,10]

# 1번째부터 8번째까지
# 인덱스 0부터7까지
t1=0
for i in range(8): #0~7
    t1= t1 + nums[i]
print(t1)

#1번째부터 3번째까지
t2=0
for i in range(3): #0~7
    t2= t2 + nums[i]
print(t2)
```

```python
nums=[1,3,5,7,9,10,20,10]
#누적합 0부터 시작
prefix=[0]

for n in nums:
    prefix.append(prefix[-1]+n) #prefix의 마지막 값에 새로들어가는 값을 더한 값을 리스트에 넣는다.
print(prefix)                   #누적합의 크기는 데이터+1 / 0때문에
```

```python
# 1부터 8끝까지 (0,7)
prefix[8]
```

```python
# 1부터 7까지 (0,6)
prefix[7]
```

```python
#원본의 5번째 7번째까지의 합 (인덱스4,6)
s=5
e=7
prefix[e]-prefix[s-1]
```

```python
#2부터5까지의 합
prefix[5] - prefix[1]
```
