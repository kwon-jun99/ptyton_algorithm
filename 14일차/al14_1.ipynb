```python
#슬라이딩 윈도우에 사용될 리스트
prices=[]

for r in sales:
    if isinstance(r["판매액"],int):
        prices.append(r["판매액"])

prices

window=3
# [0:3] [1,4] [2,5]
for i in range(0, len(prices)- (window-1)):
    total=sum(prices[i:i+window])
    avg=round(total/window,3)
    #print(avg)
    print(f"{i+1}~{i+window} 평균판매액:{avg}원")
```

```python
result={}

for r in sales:
    p_name=r["상품"]
    price=r["판매액"]
    total_cnt=r["수량"]
    if p_name not in result:
        result[p_name]={"합계":0,"건수":0,"최대":price,"최소":price,"총수량":0}
    info=result[p_name]
    info["합계"]+= price
    info["건수"]+= 1
    info["총수량"]+= total_cnt

    #최대/최소
    if info["최대"] <price:
        info["최대"]=price

    if info["최소"]>price:
        info["최소"]=price

result
```
