```python
# 2.홀수와 짝수 판별하기 (if~else)

try:
    num=int(input("숫자를 입력하세요"))

    if num%2 == 0:  
        print("짝수입니다")
    else:
        print("홀수입니다")
except ValueError as e:     # 사용자가 숫자가 아닌 문자를 입력할 수 있음. try~except로 예외경우 대책
    print("숫자만 입력하세요")
    # print(e)
```

```python
# 3. 양수,음수,0 판별하기(if~elif~else)
try:                                     #try~except
    num=int(input("숫자를 입력하세요"))     #숫자를 입력한다.

    if num>0:                            #양수면 "양수입니다" 출력
        print("양수입니다")
    elif num<0:                          #음수면 "음수입니다" 출력
        print("음수입니다")               
    else:                                #둘 다 아니면 "0"입니다 출력
        print("0입니다")                    
except ValueError as e:                  #숫자가 아닌 값 입력시(ValueError)"숫자만 입력하세요"출력
    print("숫자만 입력하세요") 
```

```python
# 5.1~50 사이의 5의 배수 출력

number=range(1,51)      #숫자 범위 정의
for i in number:        #모든 숫자 반복해서 5의 배수 찾기
    if i%5==0:          #5로 나눠서 나머지가 0이면 5의 배수
        print(i)
```

```python
# 6.1~10 사이 홀수의 합 구하기

hol=[]                      #홀수를 넣을 리스트 만들기

num1= range(1,11)           #범수 리스트 만들기
for h in num1:
    if h%2==1:              #2로 나눠서 나머지가 1이면 홀수
        hol.append(h)       #홀수인 값을 hol리스트에 넣는다

sum(hol)                    #hol리스트 안에 있는 값들을 더한다
```
