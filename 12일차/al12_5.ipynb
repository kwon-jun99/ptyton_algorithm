```python
from collections import deque

dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10)

print(dq)
```

```python
from collections import deque

dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)
```

```python
from collections import deque

dq = deque([10, 20, 30])
x = dq.popleft()

print(x)
print(dq)
```

```python
# 스텍이 필요한 경우 효율성 때문에 deque를 사용하지 않는다
# 큐 필요한 경우 효율성0 -> deque 사용
#덱 필요한 경우 deque 사용
#자료구조의 회전 시계방향, 반시계 양방향 삽입/삭제 가능
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) # deque([5, 1, 2, 3, 4])
```

```python
#최근 방분 기록 n개 저장

from collections import deque

history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
from collections import deque

cards = deque([1, 2, 3, 4, 5])

while len(cards) > 1:
    print("버림 :", cards.popleft())
    cards.append(cards.popleft()) #cards.rotate(-1)
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
