```python
scores = [10, 20, 30, 40, 50, 60]

k = 3 # window size

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[:k]) #scores[0:3]

#초기값에서 1칸을 이동 -> K까지
# window + scores[k]- scores[k-k]
# window + scores[k+1]- scores[(k+1)-k]

print(window)

for i in range(k, len(scores)): # range(3,6) 3칸 이동 / 원도우크기 부터 1칸식 이동하며 윈도우안의 합을 연산
    window = window + scores[i] - scores[i-k]
    #window+scores[3]-scores[3-3] -> 90
    #window+scores[4]-scores[4-3] -> 120
    #window+scores[5]-scores[5-3] -> 150
    print(window)
```
